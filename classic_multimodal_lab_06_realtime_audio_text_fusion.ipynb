```python
!pip install -q torchaudio librosa soundfile

import torch, requests, soundfile as sf
from transformers import pipeline

# Download a short public-domain audio sample
url = "https://huggingface.co/datasets/Narsil/asr_dummy/resolve/main/mlk.flac"
r = requests.get(url)
open("sample_audio.flac", "wb").write(r.content)

# Define pipelines with explicit models and devices
device_id = 0 if torch.cuda.is_available() else -1

asr = pipeline(
    "automatic-speech-recognition",
    model="openai/whisper-small",
    device=device_id
)

sentiment = pipeline(
    "sentiment-analysis",
    model="distilbert/distilbert-base-uncased-finetuned-sst-2-english",
    device=device_id
)

# Transcribe and analyze
result = asr("sample_audio.flac", generate_kwargs={"language": "en", "task": "transcribe"})
text = result["text"]
print("Recognized Speech:", text)

sent = sentiment(text)[0]
print("Sentiment:", sent)
```
